```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread('../data/AR-MOT/images/000001.jpg')#[159:-159, 159:-159]
```

```python
label = """
1,1,1422,2952,230,168,1
1,1,1687,3131,918,591,1
1,1,791,1711,145,141,1
1,1,1102,1636,200,161,1
1,1,3506,1695,212,137,1
1,1,624,1586,97,102,1
1,1,3202,971,358,258,1
"""
```

```python
bboxes = []
# label = open('../data/AR-MOT/labels_with_ids/000001.txt', 'r')
for line in label.strip().split('\n'):
    _, _, x, y, width, height, _ = map(float, line.split(','))
    # Convert YOLO format (relative to image size) into pixel coordinates
    # x_center = x_center * w
    # y_center = y_center * h
    # width = width * w
    # height = height * h
    # Calculate corner points from the center (YOLO gives center point)
    # x1 = int(x_center - width / 2)
    # y2 = int(y_center + height / 2)
    bboxes.append((x, y, width, height))
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray')

for bbox in bboxes:
    x, y, w, h = bbox
    print(bbox)
    plt.gca().add_patch(plt.Rectangle((x, y), w, h, edgecolor='red', facecolor='none', linewidth=2))

plt.show()
```
